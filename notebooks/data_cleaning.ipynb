```python
df = df[['Track Name', 'Artist Name(s)',
       'Release Date', 'Duration (ms)', 'Popularity', 'Explicit',
       'Added At', 'Genres', 'Danceability', 'Energy', 'Key',
       'Loudness', 'Mode', 'Speechiness', 'Acousticness', 'Instrumentalness',
       'Liveness', 'Valence', 'Tempo', 'Time Signature', 'hit_year']].copy()
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'])
```

```python
df['Release Year'] = df['Release Date'].dt.year
```

```python
df['Release Year']
```

```python
df['Hit'] = (df['Popularity']>=70).astype(int)
```

```python
#changing columns names
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ','_')
    .str.replace(r'[^0-9a-z_]', '', regex=True)
    
)
```

```python
#checking duplicates
df.query('track_name == "positions"')
```

```python
df = df.loc[~df.duplicated(subset=['track_name', 'artist_names', 'explicit'])].reset_index(drop=True).copy()
```
