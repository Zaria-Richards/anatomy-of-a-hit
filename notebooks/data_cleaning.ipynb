```python
# library imports
import pandas as pd 
```

```python
#importing and view csv
df = pd.read_csv(r"C:\Users\Owner\OneDrive\coding\what-makes-a-hit\data\spotifydataset.csv")
df.head()
```

```python
#dropping duplicates
df = df.drop_duplicates()

df.head()
```

```python
#looking at columns of data set
df.columns
```

```python
#removing un important columns
df = df.drop(columns=['Unnamed: 0', 'artist_url','followers','artist_popularity'])
df
```

```python
#converting release date columns to proper date time format
df['release_date'] = pd.to_datetime(df['release_date'],errors ='coerce')
df.head()
```

```python
#creating release year column
df['release_year'] = df['release_date'].dt.year
modern= df[df['release_year'].between(2020,2024)]
modern['release_year'] = modern['release_year'].astype(int)
modern.head()
```

```python
#checking released years columns shape 
print(modern['release_year'].unique())
print(modern.shape)
```

```python
#droping duplicates from full data set
modern = modern.drop_duplicates()
modern.head()
```

```python
#conkverting columns that may have been strings into numeric values 
numeric_cols = [
    'duration_ms','track_popularity', 'danceability',
    'energy', 'key', 'loudness', 'mode', 'speechiness',
    'acousticness', 'instrmentalness','liveness','valence',
    'tempo'
]

for col in numeric_cols:
    if col in modern.columns:
        modern[col] = pd.to_numeric(modern[col], errors='coerce')

modern
```

```python
#checking is null 
print(modern.isnull().sum())
```

```python
modern = modern.dropna(subset=['genres'])
modern
```

```python
modern['is_hit'] = (modern['track_popularity'] >= 70).astype(int)
modern['is_hit'].value_counts()
```
